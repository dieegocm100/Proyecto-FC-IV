### **Qué es un púlsar?**

Los pulsares son estrellas de neutrones altamente magentizadas, que giran rápidamente sobre su propio eje. Aquellos que son jóvenes, rotan considerablemente más rápido que los longevos, puesto que pierden energía con el paso del tiempo.
